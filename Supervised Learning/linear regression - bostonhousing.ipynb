This data frame contains the following columns:

CRIM : per capita crime rate by town.

ZN :
proportion of residential land zoned for lots over 25,000 sq.ft.

INDUS :
proportion of non-retail business acres per town.

CHAS :
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

NOX :
nitrogen oxides concentration (parts per 10 million).

RM :
average number of rooms per dwelling.

AGE :
proportion of owner-occupied units built prior to 1940.

DIS :
weighted mean of distances to five Boston employment centres.

RAD :
index of accessibility to radial highways.

TAX :
full-value property-tax rate per $10,000.

PTRATIO :
pupil-teacher ratio by town.

BLACK :
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

LSTAT :
lower status of the population (percent).

MEDV :
median value of owner-occupied homes in $1000s

```python
#Training the Linear Model
boston.columns
```

```python
X = boston[['CRIM', 'ZN', 'INDUS',
       'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD','TAX','PTRATIO']]
y = boston['MEDV']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=4)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

From the coefficient , we can definitely see that for every one unit increase in Nitrous Oxide concentration, there is a decrease of in the MEDV price. This could be a reason for the low prices in those regions.

```python
Predictions = lm.predict(X_test)
# Let's check through a scatter plot how they are aligned
plt.scatter(y_test,Predictions)
```

```python
#residual Histogram
sns.displot((y_test-Predictions),bins=50)
```

```python
#Regression Evaluation Metrics
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, Predictions))
print('MSE:', metrics.mean_squared_error(y_test, Predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, Predictions)))

#RMSE value is buffer ex: home median price is + or - 6 i.e. 14k - 26k. Lesser the buffer more closer is price accuracy of prediction
```
