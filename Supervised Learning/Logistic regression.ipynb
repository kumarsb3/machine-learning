```python
# Import Libraries
import pandas as pd
```

```python
# Read Data
data = pd.read_csv("hr_analytics.csv")
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data.sales=encoder.fit_transform(data.sales)
data.salary=encoder.fit_transform(data.salary)
```

```python
data.head()
```

```python
y=data.left
X=data.drop(['left'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y , test_size = 0.2, random_state=1)
```

```python
# Using Linear Model and select Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=10,solver='liblinear')
#model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
# Do Model prediction on test data
prediction = model.predict(X_test)
```

```python
prediction
```

```python
probabilities = model.predict_proba(X_test)
```

```python
probabilities
```

```python
# Use metrics to do accuracy score and understand the model accuracy
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy_score(prediction,Y_test)
```

```python
print(confusion_matrix(Y_test,prediction))
```

```python
from sklearn.svm import SVC
svcmodel=SVC(C=100)
svcmodel.fit(X_train,Y_train)
print(svcmodel.score(X_test,Y_test))
```

```python
# Do Model prediction on test data
svc_prediction = svcmodel.predict(X_test)
```

```python
print(confusion_matrix(Y_test,svc_prediction))
```

```python
#training and testing are done on entirely different machines
#training data need to be exported as pickle file and imported into test/inference machine
#inference machine need not be huge machine like training and this does the model prediction.

import pickle
```

```python
#creating sample pickle file
file = open('model_1_apr.pkl','wb')
```

```python
pickle.dump(model,file)
```

```python
file.close()
#this is end of training machine activity
```

```python
#This is start of test/inference machine activity

file = open('model_1_apr.pkl','rb')
mylogitmodel = pickle.load(file)
```

```python
mylogitmodel.score(X_test,Y_test)
```

```python
prediction = mylogitmodel.predict(X_test)
```
