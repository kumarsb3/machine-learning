
## Agentic AI
 - Pro active and adaptive rather than reactive -> anticipate needs, generate solutions and autonomously make decisions within defined parameters.


### Characteristics
 - goal oriented, autonomous, self learning capability, contextual awareness



### Benefits
 - Increased efficiency - task automation, process optimization, enhanced productivity
 - Improved decision making - predictive modeling leveraging data analysis
 - Personalized experiences - tailor experiences to individual needs and preferences leading to greater customer satisfaction  


### Agentic AI Design patterns
 - Goal decomposition
 - Reinforcement learning
 - Hierarchial planning
 - Multi agent coordination



## Agentic frameworks
 - langchain, langgraph, Agno, Crewai, Semantic kernel

 - LangChain and LangGraph are the tools that help build AI workflows.
 - LangChain is a framework designed to connect large language models into structured workflows enabling multi step reasoning and automation.
    - Use cases - Chatbots, document analysis and summarization

 - LangGraph is a framework that organizes large language models into graph based workflows enabling flexible and parallel multi step reasoning.
   - Use cases - Multi agent systems, dynamic AI pipelines and adaptive planning

 - Langgraph components
     - Nodes - python functions
     - Edges - connects 2 endpoints. Normal and Conditional edges
