#Visualization libraries - seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid", palette="muted")
import matplotlib.pyplot as plt
```

```python
data_df=pd.read_csv("Iris.csv")
```

```python
data_df.head()
```

```python
# joint plot is scatter plot between 2 variables, shows histogram on outer slide in addition to regular scatter plot graph

b=sns.jointplot(x='SepalWidthCm', y='SepalLengthCm', data=data_df, height=7,alpha=.9,color='b', marker='o')
#b=sns.jointplot(x='SepalWidthCm', y='SepalLengthCm', data=data_df)
```

```python
#seperate grids by category i.e. speices
c=sns.FacetGrid(data_df,col='Species')
c.map(plt.scatter,'SepalLengthCm','SepalWidthCm')
```

```python
# hue will add colors to each cateogry
c=sns.FacetGrid(data_df,hue='Species',height=5)
c.map(plt.scatter,'SepalLengthCm','SepalWidthCm').add_legend()
plt.show()
```

```python
#kind is type of plot, ex: lm, reg, etc..
e=sns.catplot(x="SepalLengthCm", y="SepalWidthCm",col="Species", data=data_df, kind="strip");
#e=sns.catplot(x="SepalLengthCm", y="SepalWidthCm", data=data_df, kind="swarm");
```

```python
#distribution of 1 variable w.r.t a category. Shows the spread among various categories

e=sns.violinplot(x='Species',y='SepalLengthCm',data=data_df, palette='magma')
```

```python
# loading dataset from seaborn and seaborn has lot of datasets as well
#bottom line is 0% value, middle is 50% value, top line is 100th, bottom and upper lines of box are 25% and 75% values

iris = sns.load_dataset("iris")
iris.head()
sns.boxplot(x="species", y="petal_length", data=iris, palette="magma",hue='species')
```

```python
sns.stripplot(x="species", y="petal_length", data=iris)
```

```python
# similar to scatter matrix. same row and colum graphs are area plots

iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', palette='gist_heat', dropna=True) # dropna is drop any missing values
#sns.pairplot(iris,hue="species")
```

```python
tips = sns.load_dataset("tips")

tips.head()
```

```python
# bar plot showing tip amount per each day
sns.barplot(x="day", y="tip", data=tips)
```

```python
# small circles are outliers.
sns.set_style("ticks")
sns.boxplot(x="day", y="total_bill", data=tips)
#sns.despine(left=True)
```

```python
# linear trend plot, blue line is regression line. wider buffer (blue area) is wide spread of data around regression line.
sns.lmplot(x="total_bill", y="tip", data=tips)
```

```python
# adding hue as gender shows different categories in colors
sns.lmplot(x="total_bill", y="tip", hue="sex",data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", col="smoker",data=tips)
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker",col="sex",data=tips, markers=['o','x'], palette="magma")
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex",data=tips, col="time",height=3 , palette="magma")
sns.set_style="whitegrid"
```

```python
sns.pairplot(data=tips, hue='sex', palette='Set1', x_vars=['total_bill','size'], y_vars=['tip'], height=6, aspect=.85, kind='reg')
```
