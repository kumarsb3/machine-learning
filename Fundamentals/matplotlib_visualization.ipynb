#Visualization libraries - matplotlib

```python
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#graphs come as inline within notebook orelse it would come as popup
#density of graph i.e. condensed or extended graph
plt.rcParams['figure.figsize'] = (4,3)
plt.rcParams['figure.dpi'] = 150
```

```python
# Basic line plot
import numpy as np
xvals = np.arange(0,10,0.1)
plt.plot(xvals, np.sin(xvals)) #y-axis data values
```

```python
# plot with axis
plt.plot(xvals, np.sin(xvals))
plt.xlabel('X value')
plt.ylabel('Sin(X)')
```

```python
# Multiple plots
plt.plot(xvals, np.sin(xvals))
plt.plot(xvals, np.cos(xvals))
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# changing color
plt.plot(xvals, np.sin(xvals), color = 'red')
plt.plot(xvals, np.cos(xvals), color = 'green')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# adding line style
plt.plot(xvals, np.sin(xvals), color = 'g')
plt.plot(xvals, np.cos(xvals), color = 'r', linestyle = '--')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# Histogram of random numbers

rands = np.random.normal(size=1000) # random normal distribution

plt.hist(rands,bins=30) # bins are boxes in histogram
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2) # 0 to 4.8 with 0.2 increments

plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 100

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# Scatter plot

n = 1024
X = np.random.normal(0,1,n)
#print (X)
Y = np.random.normal(0,1,n)

plt.rcParams['figure.figsize'] = (6,5)
plt.rcParams['figure.dpi'] = 100

plt.scatter(X,Y)
plt.show()
```

```python
n = 1024
X = np.random.normal(0,1,n) # [0-----1] mean =0 , std =1, generates 1024 data values between -3 and 3
Y = np.random.normal(0,1,n)
T = np.arctan2(Y,X) # arctan is inverse tan

plt.rcParams['figure.figsize'] = (7,6)
plt.rcParams['figure.dpi'] = 100

plt.scatter(X,Y, s=75, c=T, alpha=1) # s is size of data point, c is color of data point, alpha is transparency of data point
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.rcParams['figure.figsize'] = (6,4)
plt.rcParams['figure.dpi'] = 100

plt.scatter('a', 'b', c='c', s='d', data=data) # graph looks like b is linear function of a

plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(100)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)

plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 150

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(110, .025, r'$\mu=100,\ \sigma=15$') #latex representation where \mu is mean or average, \sigma is standard deviation, text printed at specific row and column
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
# multiple plots

plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 120

plt.subplot(2,1,1)  # first plot
plt.plot([x for x in range(25)], [y**2 for y in range(25)])
plt.plot([x for x in range(25)], [y for y in range(25)])
plt.title('Square Numbers')

plt.subplot(2,1,2) # second plot
plt.plot([x for x in range(25)], [y**3 for y in range(25)])
plt.plot([x for x in range(25)], [y for y in range(25)])
plt.title('Cube of Numbers')

plt.subplots_adjust(top=0.92, bottom=0.1, left=0.10, right=0.95, hspace=0.45,
                    wspace=0.5)
```

# when to use which plot

plot (x,y) - line (trend) representation

hist(x) - one variable distribution

scatter(x,y) - distribution between 2 variables

scatter matrix - distribution between more than 2 variables


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets #sklearn is library

iris_dataset = datasets.load_iris()
X = iris_dataset.data
Y = iris_dataset.target

iris_dataframe = pd.DataFrame(X, columns=iris_dataset.feature_names)

# Create a scatter matrix from the dataframe, color by y_train
grr = pd.plotting.scatter_matrix(iris_dataframe, c=Y, figsize=(8, 8), marker='o',
                                 hist_kwds={'bins': 20},s=10, alpha=0.9)
```
